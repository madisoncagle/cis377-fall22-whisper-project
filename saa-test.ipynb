```python
# import libraries
import os
import numpy as np
import torch
import pandas as pd
import whisper
import torchaudio

from tqdm import tqdm
```

```python
model = whisper.load_model("base")
```

```python
# get text from passage.txt
f = open("./speech-accent-archive/reading-passage.txt")
txt = f.readline().strip()
txt
```

```python
# get list of audio file names
for files in os.walk('./speech-accent-archive/recordings/recordings/'):
    recs = files[2]

recs
```

```python
path = "./speech-accent-archive/recordings/recordings/"
```

```python
# choose 100 random clips
import random

clips = []
for i in range(100):
    clips.append(random.choice(recs))

clips
```

```python
# get results
results = []
for clip in tqdm(clips):
    result = model.transcribe(path + clip, **dict(without_timestamps=True))["text"]
    results.append(result)

results
```

```python
from whisper.normalizers import EnglishTextNormalizer

normalizer = EnglishTextNormalizer()

data = pd.DataFrame(dict(transcription=results, reference=txt))
data["transcription"] = [normalizer(text) for text in data["transcription"]]
data["reference"] = [normalizer(text) for text in data["reference"]]
data
```
